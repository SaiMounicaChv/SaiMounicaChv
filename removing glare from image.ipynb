```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Open the image.
image = cv2.imread("C:/Users/chvsa/Desktop/Image Analysis/Assignment I/Image_with_glare.jpeg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Load the mask.
mask = cv2.imread("C:/Users/chvsa/Desktop/Image Analysis/Assignment I/Mask.jpeg")
mask = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
# plot the images
f = plt.figure(figsize=(12, 12))
f.add_subplot(1, 2, 1)
plt.imshow(image)
f.add_subplot(1, 2, 2)
plt.imshow(mask)
plt.show(block=True)
```

```python
# Inpaint.
dst = cv2.inpaint(image,mask, 3, cv2.INPAINT_TELEA)
f = plt.figure(figsize=(12, 12))
f.add_subplot(1, 2, 1)
plt.imshow(dst)
```
